```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df=pd.read_csv("/home/vivek/Desktop/AQI_LATEST.csv")
print(df)
df.head(10)
```

1. Which state  has the highest average pollutant level?

```python
avg_pollutant=df.groupby("state")["pollutant_avg"].mean().sort_values(ascending=False)
print(avg_pollutant)
```

```python
plt.figure(figsize=(8,6))
sns.barplot(data=avg_pollutant,palette="dark")
plt.title("State Highest pollutant average level",fontweight="bold",color="Teal")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

2. which pollutant has the highest average value across all stations?

```python
avg=df.groupby("pollutant_id")["pollutant_avg"].mean().sort_values(ascending=False)
print(avg)
```

```python
plt.figure(figsize=(6,6))
sns.barplot(data=avg,palette="coolwarm")
plt.title("Pollutant highest Average Value",fontweight="bold",color="Purple")
plt.xlabel("Pollutant_Id")
plt.ylabel("Average value")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

3. Top 10 stations with the highest average pollution?

```python
top=df.groupby('station')["pollutant_avg"].mean().sort_values(ascending=False).head(10)
print(top)
```

```python
plt.figure(figsize=(6,4))
plt.barh(top.index,top.values,data=top,color="Teal")
plt.title("Top 10 Stations highest Average pollutant",fontweight="bold",color="red")
plt.xlabel("Average value")
plt.ylabel("Station")
plt.show()
```

4. Which city has the maximum PM2.5 level?

```python
pm25 = df[df["pollutant_id"] == "PM2.5"]

city_pm25 = pm25.groupby("city")["pollutant_avg"].max().sort_values(ascending=False).head(5)

print(city_pm25)
```

```python
plt.figure(figsize=(8,5))
sns.barplot(data=city_pm25,palette="muted")
plt.title("Top 5 city PM2.5 level",fontweight="bold",color="Green")
plt.tight_layout()
plt.show()
```

5. compare the average pollution levels of diffeent cities.

```python
cities=df.groupby('city')["pollutant_avg"].mean().sort_values(ascending=False).head(20)
print(cities)
```

```python
plt.figure(figsize=(8,6))
cities.plot(kind='bar')
plt.title("(20)Average pollution level of different cities",fontweight="bold",color="blue")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

6. Which pollutant is monitored the most?

```python
count=df["pollutant_id"].value_counts()
print(count)
```

```python
plt.figure(figsize=(5,5))
sns.barplot(data=count,palette="deep")
plt.title("Monitoring of Pollutants",fontweight="bold",color="teal")
plt.show()
```

```python
plt.figure(figsize=(5,4))
plt.pie(x=count.values,labels=count.index,autopct="%1.1f%%",startangle=90,shadow=True)
plt.title("Pollutant Distribution",fontweight="bold",color="Teal")
plt.tight_layout()
plt.show()
```

7. Which stations have missing pollution data?

```python
missing=df[df['pollutant_avg'].isnull()]

missing_station=missing.groupby('station').size().sort_values(ascending=False).head(10)
print(missing_station)
```

```python
plt.figure(figsize=(6,6))
sns.barplot(data=missing_station,palette="Set2")
plt.title("Top stations with missing values",fontweight="bold",color="green")
plt.xticks(rotation=90)
plt.show()
```

8. Find the difference between maximum and minimum pollution values for each pollutant.
find which pollutant has the largest variation

```python
df["difference"]=df["pollutant_max"]-df["pollutant_min"]
print(df["difference"])

variation = (df.groupby("pollutant_id")["difference"].mean().sort_values(ascending=False))

print(variation)
```

```python
plt.figure(figsize=(6,4))
sns.barplot(data=variation,palette="rocket_r")
plt.title("Variation of pollutant",fontweight="bold",color="Green")
plt.tight_layout()
plt.show()
```

9. Which state monitors the greatest variety of pollutants ? 
   count the number of unique pollutant id in each state.

```python
high=df.groupby("state")["pollutant_id"].nunique().sort_values(ascending=False)
print(high)
```

```python
plt.figure(figsize=(8,4))
sns.barplot(data=high,palette="deep")
plt.title("State monitors greatest variety of pollutions",fontweight="bold",color="purple")
plt.ylabel("count")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
