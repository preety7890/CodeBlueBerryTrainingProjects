1. WHICH COUNTRY HAS THE HIGHEST TOTAL CO2 EMISSIONS?

```python
highest=df.groupby('country')["co2_emissions"].sum().sort_values(ascending=False).head(20)
print(highest)
```

```python
plt.figure(figsize=(8,4))
sns.barplot(data=highest,palette="coolwarm")
plt.title("Top 20 Country with Highest CO2 Emission",fontweight="bold",color="green")
plt.xticks(rotation=90)
plt.show()
```

2. WHAT IS THE AVERAGE GLOBAL TEMPERATURE ANOMALY FOR EACH YEAR?

```python
temp=df.groupby('year')["temp_anomaly_global"].mean().sort_values(ascending=False)
```

```python
plt.figure(figsize=(10,6))
sns.lineplot(data=temp,marker="o",color="green")
plt.grid(True)
plt.title("AVERAGE GLOBAL TEMPERATURE EACH YEAR",fontweight="bold",color="Brown")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

3. WHICH 10 COUNTRIES HAVE THE HIGHEST CO2 EMISSIONS

```python
top10=df.groupby('country', as_index=False)['co2_emissions'].max().sort_values(by='co2_emissions', ascending=False).head(10)
print(top10)
```

```python
plt.figure(figsize=(12,6))

sns.set_style("darkgrid")

ax = sns.barplot(
    data=top10,
    y='co2_emissions',
    x='country',
    palette='Spectral'
)

for i in ax.containers:
    ax.bar_label(i, fmt='%.1f', padding=3)

plt.title('Top 10 Countries by CO₂ Emissions',fontsize=16, fontweight='bold', color='darkblue')
plt.tight_layout()
plt.show()
```

4. WHICH CONTINENT HAS THE HIGHEST AVERAGE CO2 EMISSIONS?


```python
high=df.groupby('continent')['co2_emissions'].mean().sort_values(ascending=False)
print(high)
```

```python
sns.set_style("whitegrid")
plt.figure(figsize=(8,4))
ax = sns.barplot(
    y=high.values,
    x=high.index,
    palette='mako'
)

# Add values on bars
for i in ax.containers:
    ax.bar_label(i, fmt='%.2f', padding=3)

plt.title("Average CO₂ Emissions by Continent",fontsize=16, fontweight='bold', color='darkred')
plt.ylabel("AveargeValue")
plt.tight_layout()
plt.show()
```

5. HOW HAS ATMOSPHORIC CO2 CONCENTRATION(PPM)CHANGE OVER THE YEARS?

```python
co2_year = df.groupby('year')['co2_concentration_ppm'].mean().reset_index()
```

```python
plt.figure(figsize=(12,6))
sns.lineplot(
    data=co2_year,
    x='year',
    y='co2_concentration_ppm',
    marker='o',
    color='crimson',
    linewidth=2
)

plt.title('Atmospheric CO₂ Concentration (PPM) Change Over Years',fontsize=16,fontweight='bold')
plt.ylabel('Average CO₂ Concentration (PPM)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

6. WHICH COUNTRY HAVE THE HIGHEST PERCENTAGE OF RENEWABLE ENERGY?


```python
highest_renewable = (
    df.groupby('country')['share_renewables_pct']
      .max()
      .sort_values(ascending=False)
      .head(5)
)

print(highest_renewable)
```

```python
plt.figure(figsize=(6,4))

colors = sns.color_palette("viridis", len(highest_renewable))

plt.pie(
    highest_renewable.values,
    labels=highest_renewable.index,
    autopct='%1.1f%%',
    startangle=140,
    colors=colors,
    explode=[0.08, 0, 0, 0, 0],
    shadow=True,
    textprops={'fontsize': 11}
)

plt.title('Top 5 Countries with Highest Renewable Energy Share (%)',fontsize=16,fontweight='bold',color='darkgreen'
)

plt.axis('equal')
plt.show()
```

7. IS THERE RELATIONSHIP BETWEEN  CO2 CONCENTRATION AND GLOBAL TEMPARTURE ANOMALY?

```python
plt.figure(figsize=(8,6))

sns.set_style("whitegrid")

sns.scatterplot(
    data=df,
    x='co2_concentration_ppm',
    y='temp_anomaly_global',
    color='crimson',
    hue='continent',
    alpha=0.6,
    s=80
)

plt.title('Relationship Between CO₂ Concentration and Global Temperature Anomaly by Continent',fontweight='bold')

plt.xlabel('CO₂ Concentration (PPM)')
plt.ylabel('Global Temperature Anomaly (°C)')

plt.tight_layout()
plt.show()
```

8. WHAT IS THE AVERAGE CLIMATE RISK SCORE FOR EACH CONTINENT?

```python
climate_risk = (
    df.groupby('continent')['climate_risk_score']
      .mean()
      .sort_values(ascending=False)
)

print(climate_risk)
```

```python
plt.figure(figsize=(8,6))
sns.set_style("whitegrid")
sns.barplot(
    x=climate_risk.index,
    y=climate_risk.values,
    palette='Spectral'
)

plt.title('Average Climate Risk Score by Continent',fontweight='bold',color='brown')

plt.xlabel('Continent')
plt.ylabel('Average Climate Risk Score')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

10. WHICH VARIABLE ARE MOST CORRELATED WITH GLOBAL TEMPERATURE

```python
corr = df.corr(numeric_only=True)['temp_anomaly_global'].sort_values(ascending=False).head(10)

top_corr = corr.drop('temp_anomaly_global').head(10)
print(top_corr)
```

```python
plt.figure(figsize=(10,6))

sns.barplot(
    x=top_corr.values,
    y=top_corr.index,
    palette='coolwarm'
)

plt.title('Top 10 Variables Correlated with Global Temperature Anomaly',
          fontsize=16,
          fontweight='bold')

plt.xlabel('Correlation Coefficient')
plt.ylabel('Variables')

plt.axvline(0, color='black', linewidth=1)

plt.tight_layout()
plt.show()
```
