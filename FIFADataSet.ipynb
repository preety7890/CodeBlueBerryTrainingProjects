1. Which Continent has the highest number of teams?

```python
highest_continent=df["continent"].value_counts()
print(highest_continent)
```

```python
plt.figure(figsize=(6,6))
sns.barplot(data=highest_continent,palette="deep")
plt.title("Continent  highest number of teams",fontweight="bold",color="Brown")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

2. How many host teams are there in the dataset? (Host vs Non-Host)
  

```python
# Host teams
host_count = df["is_host"].sum()

# Non-host teams
non_host_count = (df["is_host"] == 0).sum()

print("Host teams:", host_count)
print("Non-Host teams:", non_host_count)
```

```python
plt.figure(figsize=(5,4))
plt.pie(
    [host_count, non_host_count],
    labels=["Host", "Non-Host"],
    autopct="%1.1f%%"
)

plt.title("Host vs Non-Host Teams",fontweight="bold",color="red")
plt.show()
```

3. What is the average FIFA ranking of teams from each continent?

```python
avg_ranking=df.groupby("continent")["fifa_rank_pre_tournament"].mean().sort_values(ascending=False)
print(avg_ranking)
```

```python
plt.figure(figsize=(5,5))
sns.barplot(data=avg_ranking,palette="muted")
plt.title("Average Ranking of teams each continent",fontweight="bold",color="purple")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

4. What are the top 10 teams with the highest squad market value?

```python
top10 = df[["team", "squad_total_market_value_eur"]].sort_values(by="squad_total_market_value_eur",ascending=False).head(10)

print(top10)
```

```python

plt.figure(figsize=(10,6))
sns.barplot(x="team",y="squad_total_market_value_eur",data=top10,palette="dark")

plt.title("Top 10 Teams by Squad Market Value")
plt.xlabel("Team")
plt.ylabel("Squad Market Value (EUR)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

5. Is there relationship between FIFA ranking and FIFA points?

```python
plt.figure(figsize=(5,4))
sns.scatterplot(x="fifa_points_pre_tournament",y="fifa_rank_pre_tournament",data=df,color="brown")
plt.title("Relationship between FIFA ranking and FIFA points",fontweight="bold",color="green")
plt.tight_layout()
plt.show()
```

6. Which team scored the most goals in last 4 years?

```python
top_goal=df[["team","goals_scored_last_4y"]].sort_values(by="goals_scored_last_4y",ascending=False).head(10)
print(top_goal)
```

```python
plt.figure(figsize=(8,8))
sns.barplot(x="team",y="goals_scored_last_4y",data=top_goal,palette="Set2")
plt.title("Top 10 teams scored most goals last 4 year",fontweight="bold",color="red")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

7. Compares goals scored vs goals received for each continent.

```python
comparison = df.groupby("continent")[["goals_scored_last_4y",
                                      "goals_received_last_4y"]].sum().reset_index()

print(comparison)
```

```python
plt.figure(figsize=(10,6))
comparison.plot(x="continent",kind="bar")

plt.title("Goals Scored vs Goals Received by Continent",color="purple",fontweight="bold")
plt.xlabel("Continent")
plt.ylabel("Goals")
plt.xticks(rotation=90)
plt.show()
```

8. Which teams have won the most world cup titles before the current tournament

```python
world_cup=df[["world_cup_titles_before","team"]].sort_values(by="world_cup_titles_before",ascending=False)
print(world_cup)
```

```python
plt.figure(figsize=(7,7))
sns.barplot(x="team",y="world_cup_titles_before",data=world_cup,palette="dark")
plt.title("Team Won most world cup",fontweight="bold",color="Brown")
plt.ylabel("No of World cup")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

9. What is the distribution of squad average age?

```python
plt.figure(figsize=(6,6))
sns.histplot(df["squad_avg_age"],kde=True,color="purple")
plt.title("Distribution of squad average age",fontweight="bold",color="red")
plt.tight_layout()
plt.show()
```
