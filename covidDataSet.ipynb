```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df1=pd.read_csv("/home/vivek/Desktop/covid.csv")
df2=pd.read_csv("/home/vivek/Desktop/covid_grouped.csv")
df3=pd.read_csv("/home/vivek/Desktop/coviddeath.csv")


display(df1.head(10))
display(df2.head(10))
display(df3.head(10))
```

```python
df1["Continent"]=df1["Continent"].fillna("unknown",inplace=True)

df1["Population"]=df1["Population"].fillna(0,inplace=True)
df1["WHO Region"]=df1["WHO Region"].fillna("unknown",inplace=True)

df1.fillna(0,inplace=True)

```

```python
df3["Flag"]=df3["Flag"].fillna("unknown",inplace=True)
df3.fillna(0,inplace=True)
```

1.  Find the number of row and columns each dataset? 

```python
#print("df1 shape:", df1.shape)
#print("df2 shape:", df2.shape)
#print("df3 shape:", df3.shape)

data={
    "df1":df1,
    "df2":df2,
    "df3":df3
}

for name, df in data.items():
    print(f"Number of rows and columns in {name} is:",{df.shape})
```

2. Find the top 10 countries with the highest confirmed cases..

```python
top10=df2.groupby("Country/Region")["Confirmed"].max().sort_values(ascending=False).head(10).reset_index(name="Confirmed_case")
print(top10)
```

```python
plt.figure(figsize=(6,4))
sns.barplot(x="Country/Region",y="Confirmed_case",data=top10,palette="mako")
plt.title("Top 10 Countries with confirmed cases",fontweight="bold",color="Purple")
plt.xticks(rotation=90)
plt.show()
```

3. Which country has the highest number of deaths?

```python
highest_death=df1[["Country/Region","TotalDeaths"]].sort_values(by="TotalDeaths",ascending=False).head(20)
print(highest_death)
```

```python
plt.figure(figsize=(10,8))
sns.barplot(x="Country/Region",y="TotalDeaths",data=highest_death,palette="coolwarm")
plt.title("Top 20 countries Highest Deaths",fontweight="bold",color="Green")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

4. Find the total number of confirmed ,deaths and recovered cases worldwide.

```python
latest=df2[df2["Date"]==df2["Date"].max()]
```

```python
totals=latest[["Confirmed","Deaths","Recovered"]].sum()
print(totals)
```

```python
plt.figure(figsize=(4,4))
plt.bar(totals.index,totals.values)
plt.title("World Wide total cases",fontweight="bold")
plt.show()
```

5. How many unique WHO Regions are present in datset?

```python
count=df2["WHO Region"].value_counts()
print(count)
plt.figure(figsize=(6,4))

plt.pie(
    count.values,
    labels=count.index,
    autopct="%1.1f%%"
)

plt.title("Percentage Distribution of WHO Regions",fontweight="bold",color="Teal")
plt.ylabel("")
plt.show()
```

plot a line chart showing confirmed cases overtime.

```python
df2["Date"]=pd.to_datetime(df2["Date"])
```

```python
dateall=(df2.groupby("Date")["Confirmed"].sum())

plt.figure(figsize=(12,5))
sns.lineplot(x=dateall.index,y=dateall.values,marker="o",color="Brown")
plt.grid(True)
plt.title("Confirmed cases overtime",fontweight="bold",color="Purple")
plt.xlabel("Date")
plt.ylabel("confirmed cases")
plt.show()
```

plot a histogram of confirmed cases
