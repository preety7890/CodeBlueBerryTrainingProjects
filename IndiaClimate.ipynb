3. IDENTIFY THE HOTTEST AND COLDEST YEAR IN DATASETS BASED ON AVERAGE ANNUAL TEMPAERTURE .

```python
year_temp = df.groupby("year")["avg_temp_c"].mean().reset_index()
hottest = year_temp.loc[year_temp["avg_temp_c"].idxmax()]
coldest = year_temp.loc[year_temp["avg_temp_c"].idxmin()]

print("Hottest Year:")
print(hottest)

print("\nColdest Year:")
print(coldest)
```

```python
result = pd.DataFrame([coldest, hottest])

plt.figure(figsize=(6,4))

sns.barplot(
    data=result,
    x="year",
    y="avg_temp_c",
    palette='dark',
    width=0.3
)

plt.title("Coldest and Hottest Year")
plt.xlabel("Year")
plt.ylabel("Average Temperature (°C)")

plt.show()
```

4. COMPARE THE AVERAGE TEMPERATURE HUMIDITY,AND RAINFALL ACROSS DIFFERENT SEASONS 

```python
season_avg = df.groupby("season")[["avg_temp_c", "humidity_pct", "rainfall_mm"]].mean().reset_index()

print(season_avg)
```

5. FIND THE TOP 10 STATE WITH THE HIGHEST NUMBER OF DROUGHTS OCCURRENCES DURING THE STUDY PERIOD.

```python
drought = df[df["drought_flag"] == 1]
top10 = (drought.groupby("state")
                .size()
                .sort_values(ascending=False)
                .head(10)
                .reset_index(name="Drought_Count"))

plt.figure(figsize=(10,5))

sns.barplot(
    data=top10,
    x="Drought_Count",
    y="state",
    palette="Reds_r"
)

plt.title("Top 10 States with Highest Drought Occurrences")
plt.xlabel("Number of Droughts")
plt.ylabel("State")

plt.show()
```

10. CREATE A STATE WISE CLIMATE SUMMARY REPORT SHOWING:
- AVERAGE TEMPERATURE
- TOTAL RAINFALL
- AVERAGE HUMIDITY 
- TOTAL SUNSHINE HOURS
- NUMBER OF DROUGHT EVENTS
- NUMNBER OF EL NINO RECORDS
RANK THE STATES BASED ON OVERALL CLIMATE SEVERITY.

```python
state_summary = df.groupby('state').agg({
    'avg_temp_c': 'mean',
    'rainfall_mm': 'sum',
    'humidity_pct': 'mean',
    'sunshine_hrs': 'sum',
    'drought_flag': 'sum',
    'el_nino_flag': 'sum'
})

state_summary
```

```python
state_summary['climate_severity_score'] = (
    state_summary['avg_temp_c'] +
    state_summary['drought_flag'] +
    state_summary['el_nino_flag']
)

state_rank = state_summary.sort_values(
    by='climate_severity_score',
    ascending=False
)

state_rank
```

```python
plt.figure(figsize=(8,6))

sns.barplot(
    data=state_rank.reset_index(),
    x='state',
    y='climate_severity_score',
    palette='Reds_r'
)

plt.title("State Wise Climate Severity Ranking")
plt.xlabel("State")
plt.ylabel("Climate Severity Score")
plt.xticks(rotation=45)

plt.show()
```
