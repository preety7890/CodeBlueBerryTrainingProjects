```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df=pd.read_csv("/home/vivek/Desktop/TeenMentalHealth.csv")
print(df)
df.head(10)
```

1. WHAT IS THE AVERAGE DAILY SOCIAL MEDIA USAGE (IN HOURS) FOR EACH GENDER?

```python
avg=df.groupby('gender')['daily_social_media_hours'].mean()
print(avg)
```

```python
plt.figure(figsize=(4,4))
sns.barplot(data=avg,palette="dark",width=0.2)
plt.title("Average Daily Social Media Usage",color="purple")
plt.show()
```

2. WHICH SOCIAL MEDIA PLATFORM IS THE MOST POPULAR AMONG STUDENTS?

```python
count=df['platform_usage'].value_counts()
print(count)
```

```python
sns.set_style("darkgrid")
plt.figure(figsize=(4,3))

sns.barplot(
    data=count,
    palette="Set2")

plt.title("Popularity of Social Media Platforms", fontsize=16, fontweight="bold")
plt.xlabel("Platform")
plt.ylabel("Student Count")
plt.tight_layout()
plt.show()
```

3. HOW DOES ACADEMIC PERFORMANCE VARY WITH DAILY SOCIAL MEDIA USAGE?

```python
sns.set_style("whitegrid")
plt.figure(figsize=(8,4))

sns.scatterplot(
    data=df,
    x="daily_social_media_hours",
    y="academic_performance",
    hue="gender",      
    palette="Set2",
    s=100,
    alpha=0.8
)

plt.title("Academic Performance vs Daily Social Media Usage",fontweight="bold",color='brown')
plt.xlabel("Daily Social Media Usage (Hours)")
plt.ylabel("Academic Performance")
plt.legend(title="Gender")
plt.show()
```

4. WHAT IS THE AVERAGE SLEEP DURATION FOR EACH SOCIAL MEDIA ADDICTION LEVEL?

```python
sleep=df.groupby('addiction_level')['sleep_hours'].mean()
print(sleep)
```

```python

plt.figure(figsize=(7,5))

ax = sns.barplot(data=sleep,
    palette="viridis"
)

# Add values on bars
for container in ax.containers:
    ax.bar_label(container, fmt="%.2f", fontsize=6)

plt.title("Average Sleep Duration by Social Media Addiction Level",fontweight="bold")
plt.xlabel("Addiction Level")
plt.ylabel("Average Sleep Duration (Hours)")
plt.show()
```

5. WHICH AGE GROUP HAS THE HIGHEST AVERAGE STRESS LEVEL?

```python
age=df.groupby('age')['stress_level'].mean().sort_values(ascending=False)
print(age)
```

```python
plt.figure(figsize=(7,5))

ax= sns.barplot(data=age,palette="magma")

for container in ax.containers:
    ax.bar_label(container, fmt="%.2f")

plt.title("Average Stress Level by Age Group",fontweight="bold",color="Green")
plt.xlabel("Age Group")
plt.ylabel("Average Stress Level")
plt.show()
```

6. IS THERE RELATIONSHIP BETWEEN SCREEN TIME BEFORE SLEEPS AND ANXIETY LEVELS?

```python
correlation = df["screen_time_before_sleep"].corr(df["anxiety_level"])

print("Correlation:", round(correlation, 2))
```

```python

plt.figure(figsize=(8,4))

sns.scatterplot(
    data=df,
    x="screen_time_before_sleep",
    y="anxiety_level",
    color="brown",
    s=80,
    alpha=0.7
)

plt.title("Screen Time Before Sleep vs Anxiety Level",fontweight="bold")
plt.xlabel("Screen Time Before Sleep (Hours)")
plt.ylabel("Anxiety Level")
plt.show()
```

7. WHAT IS THE PERCENTAGE OF STUDENTS IN THE DATASETS ARE CLASSIFIED AS DEPRESSED?

```python
labels = ['Not Depressed', 'Depressed']
values = df['depression_label'].value_counts().sort_index()

plt.figure(figsize=(6,4))

plt.pie(
    values,
    labels=labels,
    autopct='%1.1f%%',
    colors=['lightgreen','salmon'],
    startangle=90
)

plt.title("Percentage of Students Classified as Depressed",color="blue")
plt.show()
```

8. HOW DOES PHYSICAL ACTIVITY DIFFER BETWEEN DEPRESSED VS NON DEPRESSED STUDENTS?

```python
activity=df.groupby('depression_label')['physical_activity'].mean()
print(activity)
```

```python
plt.figure(figsize=(3,3))
sns.barplot(data=activity,palette="deep",width=0.2)
plt.title("Physical Activity")
plt.show()
```

9. WHICH SOCIAL MEDIA PLATFORMS HAS THE HIGHEST AVERAGE ADDICTION LEVEL?

```python
avg_high=df.groupby('platform_usage')['addiction_level'].mean().sort_values(ascending=False)
print(avg_high)
```

```python
plt.figure(figsize=(4,4))
sns.barplot(data=avg_high,palette="dark")
plt.title("Highest Addiction Platform",color="purple")
plt.show()
```

10. WHICH FACTORS HAVE THE STRONGEST ASSCIIATION WITH DEPRESSION?

```python
corr = df.corr(numeric_only=True)['depression_label'].drop('depression_label').sort_values(key=abs, ascending=False)

print(corr)
```

```python
plt.figure(figsize=(8,4))

plt.barh(corr.index, corr.values, color='coral')

plt.title("Factors Strongly Associated with Depression",color="Purple",fontweight="bold")
plt.xlabel("Association Strength (Correlation)")
plt.ylabel("Factors")

plt.show()
```
