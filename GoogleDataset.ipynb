```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df=pd.read_csv('/home/vivek/Desktop/google_playstore_2026.csv')
print(df)
df.head(10)
```

```python
df["Installs"] = (
    df["Installs"]
    .str.replace("+", "", regex=False)
    .str.replace(",", "", regex=False)
    .astype(int)
)
```

1. WHAT ARE THE TOP 10 APP  CATEGORIES WITH THE HIGHEST NUMNBER OF APPS?

```python
df.columns
```

```python
top10=df['Category'].value_counts().sort_values(ascending=False).head(10)
print(top10)
```

2. WHAT IS THE AVERAGE APP RATING FOR EACH CATEGORY ? WHICH CATEGORY HAS THE HIGHEST AVERAGE RATING?

```python
avg_rating = df.groupby('Category')['Rating'].mean().sort_values(ascending=False).head(10)

print(avg_rating)
```

```python
plt.figure(figsize=(8, 6))

sns.barplot(
    data=avg_rating,
    palette='viridis'
)

plt.title("Average App Rating by Category")
plt.xlabel("Category")
plt.ylabel("Average Rating")
plt.xticks(rotation=90)
plt.show()
```

8. COMAPRE THE AVERAGE RATINGS OF APPS WITH AND WITHOUT IN APP PURCHASES IS THERE ANY NOTICABLE DIFFERENCES?

```python
purchase = (
    df.groupby("In-App Purchases")["Rating"]
      .mean()
)

print(purchase)
```

```python
plt.figure(figsize=(6,5))

ax = sns.barplot(
    x=purchase.index,
    y=purchase.values,
    hue=purchase.index,
    palette="coolwarm",
    legend=False
)

for container in ax.containers:
    ax.bar_label(container,fmt="%.2f")

plt.title("Average Rating by In-App Purchases")
plt.xlabel("In-App Purchases")
plt.ylabel("Average Rating")

plt.show()
```

10. WHICH CATEGORIES HAVE THE HIGHEST PERCENTAGE OF PAID APPS AND WHAT IS AVERAGE PRICE OF PAIDS APP IN EACH CATEGORY?

```python
paid = df[df["Type"] == "Paid"].groupby("Category").size()

total = df.groupby("Category").size()

paid_percent = (paid / total * 100).sort_values(ascending=False)

print(paid_percent.head(10))
```

```python
avg_price = (
    df[df["Type"]=="Paid"]
      .groupby("Category")["Price"]
      .mean()
)

print(avg_price)
```

```python
result = pd.concat(
    [
        paid_percent.rename("Paid Percentage"),
        avg_price.rename("Average Price")
    ],
    axis=1
).sort_values("Paid Percentage",ascending=False)

print(result.head(10))
```

```python
top_paid = result.head(10)

plt.figure(figsize=(10,6))

ax = sns.barplot(
    x=top_paid["Paid Percentage"],
    y=top_paid.index,
    hue=top_paid.index,
    palette="dark",
    legend=False
)

for container in ax.containers:
    ax.bar_label(container,fmt="%.1f%%")

plt.title("Top Categories by Percentage of Paid Apps")
plt.xlabel("Paid Apps (%)")
plt.ylabel("Category")

plt.show()
```
